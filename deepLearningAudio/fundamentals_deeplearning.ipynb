 Implementing  a Neuron 
  > Implementing a Artificial neuron in py
  
 

```python
inputs =  [0.5, 0.4, 0.2]
weights = [0.4, 0.7, 0.2]

import math

def sigmoid(x):
    y = 1.0 / (1 + math.exp(-x))
    return y

def activate(inputs, weights):
    h = 0
    # First calulate the hypothesis (h) or  net input 
    for x,w in zip(inputs, weights):
        h += x*w
    # Then find the activation using the sigmoid function
    return sigmoid(h)

output = activate(inputs, weights) # this is the neuron!! 
print(output)
```
